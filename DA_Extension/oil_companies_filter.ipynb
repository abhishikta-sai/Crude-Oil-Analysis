## Computing Trend

#### Method 1:

Given a time series, if the absolute difference of the start and the end value is greater than two times the standard deviation, then there is an up trend or a down trend. Else no trend.

#### Method 2:

If the pearson correlation between the timeSeries and a straight line joining the start and end value is greater than 0.6, then we have an up trend or a down trend. Else no trend.

> Note : Method 2 is in use as it generalized better



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr

#juneMonthCrudeOil = [5344,5537,5631,5495,5473,5527,5545,5401,5268,5290,5310,5287,5125,5054,5110,5124,4889,4905,4902,4963,4965,4927]
#juneMonthCrudeOil = [5036,5131,5118,5093,5012,5090,5079,5055,4875,4803,4858,4654,4629,4720,4719,4695,4678,4747,4780,4795,4709,4815]
#juneMonthCrudeOil = [4566,4510,4516,4564,4680,4732,4655,4825,4786,4790,4796,4845,4840,4852,4889,4948,4919,4905,4831,4580,4527,4500,4609]
#juneMonthCrudeOil = [4467,4409,4637,4629,4641,4593,4439,4438,4431,4403,4451,4389,4474,4451,4481,4362,4334,4281,4321,4280]
#juneMonthCrudeOil = [3606,3777,3765,3760,3630,3808,3737,3776,3713,3764,3730,3580,3340,3350,3256,3245,3015,3220,3221,3207]
juneMonthCrudeOil = [8.64,9.1,9.55,9.06,8.68,9.08,9,9.09,9,8.61,9.51,9.31,10.79,9.52,10.39,10.49,9.97,9.89,10.39,10.17,8.88,9.28,10.25,10.6,10.19,10.07,9.15,9.4,9.01,9.44,9.01,9.45,10.19,9.5,9.44,8.35,8.76,8.4,8.8,8.95,9.25,8.99,8.92,9,9,9.2,9.95,9.5,9.97,9.63,9.69,10.1,9.7,9.3,9,8.7,9.95,10.4,10,9.9,9.7,10.31,9.81,9.41,9.85,9.43,9.47,9.38,9.7,9.48,9.48,8.8,8.5,8.64,9.3,9.35,9.8,9.38,9.2,8.8,9.23,8.95,9.4,9.92,9.53,9.1,9.51,10.32,9.85,9.45,9.26,10.49,9.85,9.75,9.35,10,9.75,9.5,9.1,9.52,9.09,9.01,9.5,9.51,10,10.53,11.07,10.75,9.76,9.76,9.67,9.21,8.01,8.37,7.98,7.6,7.93,8.02,8.22,8.15,8.15,8.57,8.2,8.6,7.84,8.4,8.25,7.87,8.23,7.85,7.48,7.87,7.5,7.15,7.48,7.15,8.23,8.67,8.26,8.31,7.93,7.55,8.73,8.35,9.62,10.12,9.66,10.13,9.66,11.17,10.66,10.16,9.68,9.68,9.22,9.22,9.2,10.02,9.8,9.95,10.95,11.4,11.49,11.13,10.6,10.09,9.75,10.25,10.1,11,10.59,10.15,9.71,9.35,9.75,10,9.64,9.39,9.09,9.5,8.71,8.55,9,9.45,9.5,9.7,9.3,9.48,9.3,9,9.4,9.55,9,9.15,8.85,8.46,8.06,7.68,7.7,7.8,8.14,7.85,8.24,7.85,7.75,8.11,7.74,7.6,7.6,7.24,6.94,6.61,6.3,6.01,5.73,5.46,5.21]
juneMonthCrudeOil.reverse()

def analyseTimeSeries(timeSeriesList):
    """
    Takes time series in list format as input and returns whether
    there was an increasing/decreasing/no trend
    increasing = 1
    decreasing = -1
    no = 0
    """
    timeSeriesMean = np.mean(timeSeriesList)
    timeSeriesStd = np.std(timeSeriesList)
    #print(timeSeriesMean, timeSeriesStd)
    timeSeriesStartValue = timeSeriesList[0]
    timeSeriesEndValue = timeSeriesList[-1]
#     if ((timeSeriesEndValue >= (mean - 1 * std)) and (timeSeriesEndValue <= (mean + 1 * std))):
#         #Return No trend
#         return 0
    timeSeriesDiffValue = timeSeriesEndValue - timeSeriesStartValue
    ### Graph Variables
    x_len = len(timeSeriesList)
    # Time series plot
    plt.plot(range(x_len), timeSeriesList)
    # Mean Line
    plt.plot(range(x_len), [timeSeriesMean]*x_len, color = 'red')
    # Mean + 1 * StdDev Line
    plt.plot(range(x_len), [timeSeriesMean + 1 * timeSeriesStd]*x_len, color = 'green')
    # Mean - 1 * StdDev Line
    plt.plot(range(x_len), [timeSeriesMean - 1 * timeSeriesStd]*x_len, color = 'green')
    
    #plt.show()
    if (abs(timeSeriesDiffValue) > 2 * timeSeriesStd):
        if (timeSeriesStartValue <= timeSeriesEndValue):
            #Increase
            return 1
        else:
            #Decrease
            return -1
    else:
        #No
        return 0
    

    

def analyseTimeSeries_2(timeSeriesList):
    timeSeriesStartValue = timeSeriesList[0]
    timeSeriesEndValue = timeSeriesList[-1]
    
    slope = (timeSeriesEndValue - timeSeriesStartValue)/(len(timeSeriesList) - 1)
    
    line = []
    for i in range(1,len(timeSeriesList)+1):
        line.append(slope * (i - 1) + timeSeriesStartValue)
    ### Graph Variables
    x_len = len(timeSeriesList)
    # Time series plot
    plt.plot(range(x_len), timeSeriesList)
    # Mean Line
    plt.plot(range(x_len), line, color = 'blue')
    plt.show()
    
    correl = pearsonr(timeSeriesList, line)[0]
    
    if(correl < 0.6):
        return 0
    else:
        if(timeSeriesStartValue < timeSeriesEndValue):
            return 1
        else:
            return -1
    
    
    
#Graph code
# x_len = len(juneMonthCrudeOil)
# plt.plot(range(x_len), juneMonthCrudeOil)
# plt.plot(range(x_len), [3850]*x_len)
# plt.plot(range(x_len), [4000]*x_len)
# plt.plot(range(x_len), [3700]*x_len)
# plt.show()

print(analyseTimeSeries(juneMonthCrudeOil))
print(analyseTimeSeries_2(juneMonthCrudeOil))
    
#[4566,4510,4516,4564,4680,4732,4655,4825,4786,4790,4796,4845,4840,4852,4889,4948,4919,4905,4831,4580,4527,4500,4609]

    
```

```python
import pandas as pd
"""
Convert date to a proper format and form a database with
trend for crude oil and other comapnies for yearly, half
yearly, quarterly, monthly
"""

### Note : Open price of crude oil is being considered


#df = pd.read_csv('DA_CrudeOil.csv')

def year_month_trend_analyse(df):    
    yearlyTrend = dict()
    monthlyTrend = dict()

    months = ["01","02","03","04","05","06","07","08","09","10","11","12"]

    for year in range(2012,2019):
        #print(year)
        dfYear = df[df.Date.str.contains(str(year))]
        if dfYear.empty == True:
            yearlyTrend["y"+str(year)] = "NOT AVAILABLE"
            for month in months:
                monthlyTrend["m"+month+"/"+str(year)] = "NOT AVAILABLE"
            continue
            
        openPrices = list(dfYear["Open Price"])
        yearlyTrend["y"+str(year)] = analyseTimeSeries_2(openPrices)
        print("Yearly Trend of "+str(year)+" : ",yearlyTrend["y"+str(year)])

        for month in months:
            #print(month+"\n")
            dfMonth = dfYear[dfYear.Date.str.contains("/"+month+"/")]
            if dfMonth.empty == True:
                monthlyTrend["m"+month+"/"+str(year)] = "NOT AVAILABLE"
            else:
                monthlyOpenPrices = list(dfMonth["Open Price"])
                monthlyTrend["m"+month+"/"+str(year)] = analyseTimeSeries_2(monthlyOpenPrices)
            #print(monthlyTrend["m"+month+"/"+str(year)])
    
    return yearlyTrend, monthlyTrend
        
```

```python
#Half Yearly analysis
def half_yearly_analyse(df):
    halfYearlyTrend = dict()
    months = ["01","02","03","04","05","06","07","08","09","10","11","12"]
    for year in range(2012,2019):
        #print(year)
        dfYear = df[df.Date.str.contains(str(year))]
        if dfYear.empty == True:
            halfYearlyTrend["h01/"+str(year)] = "NOT AVAILABLE"
            halfYearlyTrend["h02/"+str(year)] = "NOT AVAILABLE"
            continue
        
        halfYear1 = []
        halfYear2 = []
        #1st half
        for month in months[:6]:
            dfMonth = dfYear[dfYear.Date.str.contains("/"+month+"/")]
            if dfMonth.empty == True:
                halfYearlyTrend["h01/"+str(year)] = "NOT AVAILABLE"
                break
            monthlyOpenPrices = list(dfMonth["Open Price"])
            halfYear1 = halfYear1 + monthlyOpenPrices

        #2nd half
        for month in months[6:]:
            dfMonth = dfYear[dfYear.Date.str.contains("/"+month+"/")]
            if dfMonth.empty == True:
                halfYearlyTrend["h02/"+str(year)] = "NOT AVAILABLE"
                break
            monthlyOpenPrices = list(dfMonth["Open Price"])
            halfYear2 = halfYear2 + monthlyOpenPrices

        print("1st half of ",year)
        if "h01/"+str(year) not in halfYearlyTrend.keys():
            halfYearlyTrend["h01/"+str(year)] = analyseTimeSeries_2(halfYear1)
        print(halfYearlyTrend["h01/"+str(year)])
        print("2nd half of ",year)
        if "h02/"+str(year) not in halfYearlyTrend.keys():
            halfYearlyTrend["h02/"+str(year)] = analyseTimeSeries_2(halfYear2)
        print(halfYearlyTrend["h02/"+str(year)])
        
    return halfYearlyTrend
        
```

```python
#Quarterly Analysis

def quarterly_analyze(df):
    quarterlyTrend = dict()
    months = ["01","02","03","04","05","06","07","08","09","10","11","12"]
    for year in range(2012,2019):
        #print(year)
        dfYear = df[df.Date.str.contains(str(year))]
        if dfYear.empty == True:
            quarterlyTrend["q01/"+str(year)] = "NOT AVAILABLE"
            quarterlyTrend["q02/"+str(year)] = "NOT AVAILABLE"
            quarterlyTrend["q03/"+str(year)] = "NOT AVAILABLE"
            quarterlyTrend["q04/"+str(year)] = "NOT AVAILABLE"
            continue
        quarter1 = []
        quarter2 = []
        quarter3 = []
        quarter4 = []

        for month in months[:3]:
            dfMonth = dfYear[dfYear.Date.str.contains("/"+month+"/")]
            if dfMonth.empty == True:
                quarterlyTrend["q01/"+str(year)] = "NOT AVAILABLE"
                break
            monthlyOpenPrices = list(dfMonth["Open Price"])
            quarter1 = quarter1 + monthlyOpenPrices

        for month in months[3:6]:
            dfMonth = dfYear[dfYear.Date.str.contains("/"+month+"/")]
            if dfMonth.empty == True:
                quarterlyTrend["q02/"+str(year)] = "NOT AVAILABLE"
                break
            monthlyOpenPrices = list(dfMonth["Open Price"])
            quarter2 = quarter2 + monthlyOpenPrices

        for month in months[6:9]:
            dfMonth = dfYear[dfYear.Date.str.contains("/"+month+"/")]
            if dfMonth.empty == True:
                quarterlyTrend["q03/"+str(year)] = "NOT AVAILABLE"
                break
            monthlyOpenPrices = list(dfMonth["Open Price"])
            quarter3 = quarter3 + monthlyOpenPrices

        for month in months[9:]:
            dfMonth = dfYear[dfYear.Date.str.contains("/"+month+"/")]
            if dfMonth.empty == True:
                quarterlyTrend["q04/"+str(year)] = "NOT AVAILABLE"
                break
            monthlyOpenPrices = list(dfMonth["Open Price"])
            quarter4 = quarter4 + monthlyOpenPrices

        print("1st quarter of ",year)
        if "q01/"+str(year) not in quarterlyTrend.keys():
            quarterlyTrend["q01/"+str(year)] = analyseTimeSeries_2(quarter1)
        print(quarterlyTrend["q01/"+str(year)])

        print("2nd quarter of ",year)
        if "q02/"+str(year) not in quarterlyTrend.keys():
            quarterlyTrend["q02/"+str(year)] = analyseTimeSeries_2(quarter2)
        print(quarterlyTrend["q02/"+str(year)])

        print("3rd quarter of ",year)
        if "q03/"+str(year) not in quarterlyTrend.keys():
            quarterlyTrend["q03/"+str(year)] = analyseTimeSeries_2(quarter3)
        print(quarterlyTrend["q03/"+str(year)])

        print("4th quarter of ",year)
        if "q04/"+str(year) not in quarterlyTrend.keys():
            quarterlyTrend["q04/"+str(year)] = analyseTimeSeries_2(quarter4)
        print(quarterlyTrend["q04/"+str(year)])
        
    return quarterlyTrend

        
    
```
