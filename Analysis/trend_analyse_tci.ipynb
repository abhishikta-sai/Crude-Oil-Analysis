```python
import pandas as pd
import datetime
from matplotlib import pyplot as plt
```

```python

df = pd.read_csv("merge_oil.csv")

dates = [datetime.date(int(i.split('-')[0]),int(i.split('-')[1]),int(i.split('-')[2])) for i in list(df["Date"])]

df["Dates"] = dates

df["Year"] = [int(i.split('-')[0]) for i in df["Date"]]
df["Month"] = [int(i.split('-')[1]) for i in df["Date"]]
df["Day"] = [int(i.split('-')[2]) for i in df["Date"]]


years = list(set([i.year for i in dates]))
years.sort()
#years

yearly_data_oil = dict()

for i in years:
    yearly_data_oil[i] = df[df["Year"]==i]
    yearly_data_oil[i] = yearly_data_oil[i][["TCI_Close","Year","Month","Day","Dates"]]
    l = list(yearly_data_oil[i]["TCI_Close"])
    l.reverse()
    yearly_data_oil[i]["TCI_Close"] = l
    
    l = list(yearly_data_oil[i]["Year"])
    l.reverse()
    yearly_data_oil[i]["Year"] = l
    
    l = list(yearly_data_oil[i]["Month"])
    l.reverse()
    yearly_data_oil[i]["Month"] = l
    
    l = list(yearly_data_oil[i]["Day"])
    l.reverse()
    yearly_data_oil[i]["Day"] = l
    
    l = list(yearly_data_oil[i]["Dates"])
    l.reverse()
    yearly_data_oil[i]["Dates"] = l
    

    
for i in years:
    print(i)
    plt.plot(yearly_data_oil[i]["Dates"],yearly_data_oil[i]["TCI_Close"])
    plt.xticks(rotation=90)
    plt.show()
    
    
#print(yearly_data_oil[2011].head())

#print(len(list(yearly_data_oil[2012]["Dates"])))

    

#getting week number for a given date
#datetime.date(2010, 6, 16).isocalendar()[1]
```

```python
half_yearly_data_oil = dict()
years = list(range(2012,2019))
for i in years:
    half_yearly_data_oil[(1,i)] = yearly_data_oil[i][yearly_data_oil[i]["Month"]<=6]
    half_yearly_data_oil[(2,i)] = yearly_data_oil[i][yearly_data_oil[i]["Month"]>6]
    
#print(half_yearly_data_oil[(2,2012)].head())    

for i in half_yearly_data_oil.keys():
    print(i)
    plt.plot(half_yearly_data_oil[i]["Dates"],half_yearly_data_oil[i]["TCI_Close"])
    plt.xticks(rotation=90)
    plt.show()

```

```python
quarter_yearly_data_oil = dict()

for i in years:
    quarter_yearly_data_oil[(1,i)] = yearly_data_oil[i][yearly_data_oil[i]["Month"]<=3]
    quarter_yearly_data_oil[(2,i)] = yearly_data_oil[i][(yearly_data_oil[i]["Month"]<=6) & (yearly_data_oil[i]["Month"]>3)]
    quarter_yearly_data_oil[(3,i)] = yearly_data_oil[i][(yearly_data_oil[i]["Month"]<=9) & (yearly_data_oil[i]["Month"]>6)]
    quarter_yearly_data_oil[(4,i)] = yearly_data_oil[i][yearly_data_oil[i]["Month"]>9]
    
for i in quarter_yearly_data_oil.keys():
    print(i)
    plt.plot(quarter_yearly_data_oil[i]["Dates"],quarter_yearly_data_oil[i]["TCI_Close"])
    plt.xticks(rotation=90)
    plt.show()
    
```
